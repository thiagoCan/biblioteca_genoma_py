```python
g={}

for i in ['A', 'T', 'C', 'G']:
    for j in ['A', 'T', 'C', 'G']:
        g[i+j] = 0
g
```

```python
a = open('bacteria1.txt')
a = a.read()
a = a.replace('\n', '')
a
```

```python
for k in range(len(a) - 1):
    g[a[k]+a[k+1]] += 1   
g
```

```python
import pandas as pd
```

```python
h = pd.Series(g)
h
```

```python
hx = h.index
hx
```

```python
hy = h.values
hy
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(hx, hy, color='#FF0000')
plt.plot(hx, hy)
plt.show()
```

```python
z = {}

for q in ['A', 'T', 'C', 'G']:
    for p in ['A', 'T', 'C', 'G']:
        z[q+p] = 0
z
```

```python
s = open('human1.txt')
s = s.read()
s = s.replace('\n', '')
s
```

```python
for a in range(len(s)-1):
    z[s[a]+s[a+1]] += 1
z
```

```python
z = pd.Series(z)
z
```

```python
zx = z.index
```

```python
zy = z.values
```

```python
plt.scatter(zx, zy, color='#000000')
plt.plot(zx, zy, color='#00FF00')
plt.show()
```
